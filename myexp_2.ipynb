```python
def cap_outlier(col):
    Q1=col.quantile(0.25)
    Q2=col.quantile(0.75)
    IQR=Q2-Q1
    lower=Q1-1.5*IQR
    upper=Q2+1.5*IQR
    return col.clip(lower, upper) 
```

```python
df["Math"]=cap_outlier(df["Math"])
df["Science"]=cap_outlier(df["Science"])
```
